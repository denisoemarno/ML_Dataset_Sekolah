#IMPORT LIBRARY

```python
import pandas as pd
import numpy as np
```

#LOAD DATA

```python
sekolah = pd.read_csv('data.csv')
sekolah
```

#Mencari Rata Rata Jumlah Atribut yang dimiliki

```python
#filter sekolah by provinsi
sekolahByProvinsi = sekolah.loc[sekolah.provinsi == 11]
sekolahByProvinsi
```

#Lab Sekolah dari Provinsi == 11

```python
params_sekolah = sekolahByProvinsi.iloc[:,36:44].mean()
params_sekolah
```

#Lab Terbanyak Sekolah SMK

```python
sekolahSMK = sekolah.loc[sekolah.bentuk == 'SMK']
labTerbanyakPerjenis = sekolahSMK.iloc[:,36:44].sum()
labTerbanyak = labTerbanyakPerjenis.idxmax()
print(labTerbanyakPerjenis)
print("\n\n")
print("{} adalah jumlah lab terbanyak".format(labTerbanyak))
```
